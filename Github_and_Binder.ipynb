##GitHub

###What is GitHub?

GitHub is an internet hosting site for software developers. It is the biggest provider for source code management and version control. In more understandable terms, software developers upload their code base to GitHub frequently while programming. It saves previous versions of their code so that developers can look back at or revert their code to a previous state. 

For this project we are just using GitHub for its file hosting. We are going to upload our notebook, our pkl file and a "requirements" text file for Binder to use.

###Creating a GitHub Account

To sign up for Github go to [github.com](https://github.com/).

Click on the Sign up button in the top right corner and then create your account.

###Creating a new GitHub Repository

In the top left of your GitHub home page, below the search bar, you should see a green button that says "New". Click on it to get to access the Create a new Repository page.

Next, name your repository and make sure that it is public. You do not need to check any of the other boxes. Then just click the green Create repository button at the bottom and you are done!

###Uploading your files to GitHub

Next we need to upload our three files. Once again, these are:


1.   Your colab notebook (.ibynb file)
2.   export.pkl
1.   requirements.txt





####Uploading your Colab Notebook

GitHub is such a foundational part of code development that there is an option to save your notebook to Github right in the file menu. 

colabgithub.jpg

It should automatically detect your signed in GitHub account. You can then click on the left menu to select the repository where you want to store this notebook. Leave the branch set to "master".

Go back to the GitHub page and double check that your notebook is now in your repository. You should be all set with the first step!

####Uploading Your Export File

Working with your export.pkl file requires downloading the file to your computer. If you are unable to do this ask your coach or your instructor to help you with this step.

Your export.pkl file is currently saved to your Google Drive. Head into your drive and search for export.pkl. Right click on that file and then select "Download" from the dropdown menu. This should download your export.pkl file into your downloads folder.

Next, head back to GitHub. In your repository page on GitHub there is an button labeled "Add file". Click on that button, navigate to where your export.file is downloaded and then select it. 

If your export.pkl file is under 25mb then this is all you need to do to have your file uploaded. However, if you have a large file then we have a little more work to do.

For large .pkl files (over 25mb) you need to download the GitHub Desktop app. You may need to seek help from your coach or instructor with this. Once the app is downloaded, open it, log in and select your repository. You can then "pull" the repository to your computer which will download everything from the repository and create a folder with your current files. 

At this point you can click on the "View the files of your repository in Explorer" button to open a local folder. Drag your export.pkl file into this folder. 

Back in the GitHub app, it should list your changed file and provide a blue button in the bottom lefthand corner that says "Commit to main". This button will update your branch locally. Then, click on the "Push origin" button on your top bar to upload your new changes to your online GitHub repository. Check your online GitHub repository to double check that export.pkl is there and you are all set.

####Creating requirements.txt

The last step is the easiest. We need our notebook to have access to three libraries - fastai, ipywidgets and voila. ipywidgets and voila provide the app features we used in the last two notebooks. We can provide access to these libraries by adding a file named "requirements.txt" to our repository that includes these libraries.

Head to your GitHub page and hit the "Add file" button we used to upload export.pkl. Select "Create new file" from the dropdown menu. In the "Name your file" input field, type requirements.txt.

This file is just three words - type


```
fastai
voila
ipywidgets
```
on lines 1 through 3. You don't need anything else! Scroll to the bottom of the page, make sure the "Commit directly to the main branch" radio button is selected and then click on the green "Commit new file" button.

This should take you back to the main page of your repository where you should now see your notebook (.ipynb file), export.pkl and requirements.txt. Our GitHub repository is now all set!


##Binder

Now we are on the home stretch! All that is left is to connect our GitHub repository to Binder.

Binder is a website designed to turn Jupyter Notebooks, like the ones we built with Google Colaboratory, into an interactive online environment. This is exactly what we need!

Head on over to [mybinder.org](https://mybinder.org).

No account or anything is required. All you need to do is select GitHub from the first menu and then paste in the web address of your GitHub repository. Next, switch the dropdown that is automatically set to "File" to "URL" instead. It will reset to "File" anytime you reload the page - this will be the most common cause for an error!

In the URL to open field, type 



```
/voila/render/name.ipynb 
```
where "name" is the name of the notebook you have saved in your GitHub repository. Then, copy the link using the clipboard button below to share with other people. When you launch your site it will provide a non-permanent web address that is not useful for sharing purposes. You can use this link to look at your app on computers, tablets or phones.

Lastly, click the orange launch button! It will take a few minutes but you should see your app in all of its glory.
